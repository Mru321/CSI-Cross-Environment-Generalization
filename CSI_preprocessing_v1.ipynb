Function to convert the complex number (String) to its amplitude (int)

```python
def safe_complex(x):
    try:
        x = str(x).strip()
        x = x.replace('+-', '-')
        x = x.replace('i', 'j')
        return complex(x)
    except:
        return 0+0j
```

```python
raw_base = "/content/drive/MyDrive/FYP/raw_data/"
processed_base = "/content/drive/MyDrive/FYP/processed_data/"
```

Creating processed files from the raw dataset files

```python
for root, dirs, files in os.walk(raw_base):
    #print("Scanning folder:", root)
    for file in files:

        if file.lower().endswith(".csv"):

            raw_file_path = os.path.join(root, file)
            #print("Reading:", raw_file_path)

            df = pd.read_csv(raw_file_path)

            # keep only CSI columns
            csi_cols = [col for col in df.columns if "csi_" in col]
            df = df[csi_cols]

            # convert complex
            df = df.applymap(safe_complex)

            # amplitude
            df = df.applymap(np.abs)

            # recreate folder structure
            relative_path = os.path.relpath(root, raw_base)
            processed_folder = os.path.join(processed_base, relative_path)
            os.makedirs(processed_folder, exist_ok=True)

            processed_file_path = os.path.join(processed_folder, file)

            df.to_csv(processed_file_path, index=False)

            #print("Saved:", processed_file_path)

            #stop after first file
            #raise SystemExit
```

Verify Folder Structure Matches

```python
missing_files = []

for root, dirs, files in os.walk(raw_base):
    for file in files:
        if file.lower().endswith(".csv"):

            raw_file = os.path.join(root, file)
            relative_path = os.path.relpath(root, raw_base)
            processed_file = os.path.join(processed_base, relative_path, file)

            if not os.path.exists(processed_file):
                missing_files.append(processed_file)

print("Missing processed files:", len(missing_files))
```

Reprocess only the missing files

```python
for processed_file in missing_files:

    # reconstruct raw file path
    relative_path = os.path.relpath(processed_file, processed_base)
    raw_file = os.path.join(raw_base, relative_path)

    if os.path.exists(raw_file):

        df = pd.read_csv(raw_file)

        csi_cols = [col for col in df.columns if "csi_" in col]
        df = df[csi_cols]

        df = df.applymap(safe_complex)
        df = df.applymap(np.abs)

        os.makedirs(os.path.dirname(processed_file), exist_ok=True)

        df.to_csv(processed_file, index=False)

print("Missing files processed.")
```

Check If Any Cell Contains "i" (Complex String)

```python
import os
import pandas as pd

processed_base = "/content/drive/MyDrive/FYP/processed_data/"

complex_found = False

for root, dirs, files in os.walk(processed_base):
    for file in files:
        if file.endswith(".csv"):
            path = os.path.join(root, file)
            df = pd.read_csv(path)

            if df.astype(str).apply(lambda x: x.str.contains("i")).any().any():
                print("Complex number found in:", path)
                complex_found = True
                break

    if complex_found:
        break

if not complex_found:
    print("✅ No complex numbers found in processed dataset.")
```

Fast Dataset Validation Script

```python
processed_base = "/content/drive/MyDrive/FYP/processed_data/"

files_checked = 0
datatype_issue = False
nan_issue = False

for root, dirs, files in os.walk(processed_base):
    for file in files:
        if file.endswith(".csv"):

            path = os.path.join(root, file)

            # read small sample instead of full file
            df = pd.read_csv(path, nrows=50)

            files_checked += 1

            # check datatype
            if "object" in df.dtypes.astype(str).values:
                print("❌ Non-numeric values detected in:", path)
                datatype_issue = True
                break

            # check NaN
            if df.isnull().values.any():
                print("❌ NaN values detected in:", path)
                nan_issue = True
                break

    if datatype_issue or nan_issue:
        break


print("Files checked:", files_checked)

if not datatype_issue and not nan_issue:
    print("✅ Dataset validation successful")
    print("All values are numeric amplitudes")
```

Creating new features for classifying the data into los and nlos and merging all files into Final_dataset.csv

```python
import os
import pandas as pd

processed_base = "/content/drive/MyDrive/FYP/processed_data/"
output_file = "/content/drive/MyDrive/FYP/final_dataset.csv"

first_file = True

for root, dirs, files in os.walk(processed_base):

    for file in files:

        if file.endswith(".csv"):

            file_path = os.path.join(root, file)

            df = pd.read_csv(file_path)

            if "env1" in root:
                label = 0
                env = "env1"

            elif "env2" in root:
                label = 0
                env = "env2"

            elif "env3" in root:
                label = 1
                env = "env3"

            df["label"] = label
            df["environment"] = env

            if first_file:
                df.to_csv(output_file, index=False, mode="w")
                first_file = False
            else:
                df.to_csv(output_file, index=False, mode="a", header=False)

print("Dataset creation completed")
```

Process the data in chunks to count the number of data pkt according to its environment

```python
import pandas as pd

file_path = "/content/drive/MyDrive/FYP/final_dataset.csv"

env_label_check = {}

for chunk in pd.read_csv(file_path, chunksize=200000):

    counts = chunk.groupby(["environment", "label"]).size()

    for key, value in counts.items():
        env_label_check[key] = env_label_check.get(key, 0) + value

print(env_label_check)
```
